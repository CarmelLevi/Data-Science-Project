```python
Halisa= pd.read_excel(r'C:\Users\calevi.MIDDLEEAST\Halisa.xlsx')
Halisa["Neighborhood"] = pd.NaT
Halisa.replace(to_replace= np.nan, value = 1, inplace=True )
Halisa.to_excel(r'C:\Users\calevi.MIDDLEEAST\Halisa.xlsx',index=False)

Vizni= pd.read_excel(r'C:\Users\calevi.MIDDLEEAST\Vizni.xlsx')
Vizni["Neighborhood"] = pd.NaT
Vizni.replace(to_replace= np.nan, value = 2, inplace=True )
Vizni.to_excel(r'C:\Users\calevi.MIDDLEEAST\Vizni.xlsx',index=False)

HadarCarmel= pd.read_excel(r'C:\Users\calevi.MIDDLEEAST\HadarCarmel.xlsx')
HadarCarmel["Neighborhood"] = 3
HadarCarmel.to_excel(r'C:\Users\calevi.MIDDLEEAST\HadarCarmel.xlsx',index=False)

Carmelia= pd.read_excel(r'C:\Users\calevi.MIDDLEEAST\Carmelia.xlsx')
Carmelia["Neighborhood"] = 4
Carmelia.to_excel(r'C:\Users\calevi.MIDDLEEAST\Carmelia.xlsx',index=False)

Denia2= pd.read_excel(r'C:\Users\calevi.MIDDLEEAST\Denia2.xlsx')
Denia2["Neighborhood"] = 5
Denia2.to_excel(r'C:\Users\calevi.MIDDLEEAST\Denia2.xlsx',index=False)

Denia= pd.read_excel(r'C:\Users\calevi.MIDDLEEAST\Denia.xlsx')
Denia["Neighborhood"] = 6
Denia.to_excel(r'C:\Users\calevi.MIDDLEEAST\Denia.xlsx',index=False)

Galoyot= pd.read_excel(r'C:\Users\calevi.MIDDLEEAST\Galoyot.xlsx')
Galoyot["Neighborhood"] = pd.NaT
Galoyot.replace(to_replace= np.nan, value = 7, inplace=True )
Galoyot.to_excel(r'C:\Users\calevi.MIDDLEEAST\Galoyot.xlsx',index=False)

Amos= pd.read_excel(r'C:\Users\calevi.MIDDLEEAST\Amos.xlsx')
Amos["Neighborhood"] = pd.NaT
Amos.replace(to_replace= np.nan, value = 8, inplace=True )
Amos.to_excel(r'C:\Users\calevi.MIDDLEEAST\Amos.xlsx',index=False)


#Combine all neighborhoods
names = ['Halisa','Vizni','Hadar Carmel','Carmelia','Savyoni Denia','Denia', 'Galoyot', 'Amos']
AllNeghiborhoods = [Halisa,Vizni,HadarCarmel,Carmelia,Denia2,Denia, Galoyot, Amos]
new_excel = pd.concat(AllNeghiborhoods)
new_excel.to_excel(r'C:\Users\calevi.MIDDLEEAST\AllNeighborhoods.xlsx',index=False)

#Drop and change unrelated characters/data
def Drop_Change(df):
    df=df.replace(to_replace='ק', value=0)
    df=df.replace('-',np.NaN)
    df=df.dropna()
    df = df.drop(columns='Address')
    df['Price to SQ']=df['Price to SQ'].str.replace(',','')
    df['Price to SQ']=df['Price to SQ'].str.replace('₪','')
    df['Price']=df['Price'].str.replace('₪','')
    df['Price']=df['Price'].str.replace('א׳','')
    df['Price']=df['Price'].str.replace('מ׳','')
    return (df)

#Change columns to numeric
def Numeric(df):
    df['Price to SQ']=pd.to_numeric(df['Price to SQ'])
    df['SQ']=pd.to_numeric(df['SQ'])
    df['Floor']=pd.to_numeric(df['Floor'])
    df['Rooms']=pd.to_numeric(df['Rooms'])
    df['Price']=pd.to_numeric(df['Price'])
    df['Price']=df['Price to SQ']*df['SQ']
    return(df)

#Split dates
def Split(df):
    df["Day"] = pd.NaT
    df["Month"] = pd.NaT
    df["Year"] = pd.NaT
    df['Transaction date']=pd.to_datetime(df['Transaction date'],format='%d.%m.%Y')
    df['Year']=df['Transaction date'].dt.year
    df['Month']=df['Transaction date'].dt.month
    df['Day']=df['Transaction date'].dt.day
    df=df.drop(['Transaction date'], axis=1)
    return(df)
    
#Check wheather there are missing or null values
for i in range(len(AllNeghiborhoods)):
    AllNeghiborhoods[i]=Drop_Change(AllNeghiborhoods[i])
    AllNeghiborhoods[i]=Numeric(AllNeghiborhoods[i])
    AllNeghiborhoods[i]=Split(AllNeghiborhoods[i])
    AllNeghiborhoods[i].isnull().sum()
    

new_excel=Drop_Change(new_excel)
new_excel=Numeric(new_excel)
new_excel=Split(new_excel)
```

```python
#Outliners & Standard deviation function
def outlier(df):
    for col in new_excel.columns:
        Q1 = np.percentile(df[col],25)
        Q3 = np.percentile(df[col],75)
        IQR=Q3-Q1
        df[(df[col]<Q1-(1.5*IQR))|(df[col]>Q3+1.5*IQR)] = np.nan
        #Remove outliners from df
        df=df[df[col].notna()]
        return df
```

# Machine Learning 

```python
def preparation(dataframe):
    X=dataframe.drop(['Price'], axis=1)
    X = X.drop(['Neighborhood'], axis=1)

    Y=dataframe['Price']
    return X,Y
```

```python
def Train_test_split(X,Y):
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size =0.2, random_state= 1)
    return X_train, X_test, Y_train, Y_test
```

```python
def Linear_regression(X_train, Y_train):
    model = LinearRegression()
    #Training the model with X_train
    model.fit(X_train, Y_train)
    return model
```

```python
def Predict(model, X_test):
    Y_predicted = model.predict(X_test)
    return Y_predicted
```

```python
def Calculate_Evaluation(y_test, y_predicted):
    evaluation = metrics.r2_score(y_test, y_predicted)
    return evaluation
```

```python
#Visualizing the actual prices and predicted prices 
def Actual_vs_Predicted(y_train, x_train, y_test):
    #Accuracy for prediction on training data
    training_data_prediction=model.predict(x_train)
    plt.plot(Y_test, Y_test, color='red', linewidth=2)
    plt.scatter(y_train, training_data_prediction)
    plt.xlabel("Actual prices")
    plt.ylabel("Predicted prices")
    plt.title ("Actual vs Predicted")
    plt.show()
```

```python
for i in range(len(AllNeghiborhoods)):
    X,Y = preparation(AllNeghiborhoods[i])
    X_train, X_test, Y_train, Y_test=Train_test_split(X,Y)
    model = Linear_regression(X_train, Y_train)
    Y_predicted=Predict(model, X_test)
    evaluation = Calculate_Evaluation(Y_test, Y_predicted)
    print("Evaluation for {} neighborhood:".format(names[i]), evaluation)
    Actual_vs_Predicted(Y_train, X_train, Y_test)
    
```
